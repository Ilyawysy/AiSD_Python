5) Реализовать функцию вычисления факториала с помощью хвостовой
рекурсии.

```python
c = 1
def factorial(n):
    global c
    c += 1
    if n == 0:
        return ''
    print(f'{"  " * (c - 1)}factorial({n})')
    return factorial(n - 1)
print(factorial(5))
```

6) Реализовать функцию подсчета n-го числа Фибоначчи с помощью хвостовой рекурсии.

```python
def fib(n):
    if n in (1, 2):
        return 1
    return fib(n - 1) + fib(n - 2)
 
print(fib(10))
```

7) Реализовать декоратор, который выводит на печать возвращаемые значения функции.

```python
def decor1(func):
    
    def wrapper(n):
        result = func(n)
        print(f'result of {func.__name__} is {result}')
        return result
    
    return wrapper



def decor2(f):
    
    def wrapper(n):
        print(f"Вызов функции {f.__name__} с аргументом {n}")
        return f(n)
    
    return wrapper
```

```python
@decor1
def fib(n):
    if n in (1, 2):
        return 1
    return fib(n - 1) + fib(n - 2)
 
print(fib(10))
```

```python
c = 1
@decor2
def factorial(n):
    global c
    c += 1
    if n == 0:
        return ''
    print(f'{"  " * (c - 1)}factorial({n})')
    return factorial(n - 1)
print(factorial(5))
```

8) С помощью декоратора реализовать отладочный вывод работы factorial(n) как для вызовов функций, так и для возвращаемых значений.

```python
def decor(f):
    def wrapper(n):
        print(f"Вызов функции {f.__name__} с аргументом {n}")
        return f(n)
    return wrapper

c = 1
@decor
def factorial(n):
    global c
    if n == 0:
        return ''
    print(f'{"  " * (c)}factorial({n})')
    c += 1
    return factorial(n - 1)
print(factorial(4))
```

9) Модернизировать пункт 8) при помощи functools.wraps(). https://stackoverflow.com/a/309000

```python
from functools import wraps
def decor(func):
    @wraps(func)
    def wrapper(n):
        print(func.__name__ + " was called")
        return func(n)
    return wrapper
c = 1
@decor
def factorial(n):
    global c
    if n == 0:
        return ''
    print(f'{"    " * (c)}factorial({n})')
    c += 1
    return factorial(n - 1)
print(factorial(7))
```

10) Реализовать декоратор, который выводит на печать имя функции и ее аргументы (на одной строке) и длительность исполнения и возвращаемые значения функции (на другой строке). Вывод на печать оформляется отступами, соответсвующими вложенности вызова функции. Проверить на запуске рекурсивной реализации получения чисел Фибоначчи.

```python
import time
def decor(func):
    def wrapper(n):
        start = time.time()
        print(f"Вызов функции {func.__name__} с аргументом {n}")
        func(n)
        end = time.time()
        print(end-start)
        return func(n)
    return wrapper


@decor
def fibonacci(n):
    if n in (0, 1):
        return 1
    return fibonacci(n - 1) + fibonacci(n - 2)
print(fibonacci(5))
```
